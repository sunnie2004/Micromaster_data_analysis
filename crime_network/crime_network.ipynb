```python
G_edge_list = list(G.edges(data=True))   #list(tuples(u,v,{dict}))
G_edge_number =  len(G_edge_list)
print(G_edge_number)
for i in range(G_edge_number):
  if G_edge_list[i][2]['weight'] < 2:    #the edge less than 2 to remove, =>2 keep the node and edges 
    G.remove_edge(G_edge_list[i][0],G_edge_list[i][1])
  if G_edge_list[i][2]['weight'] !=1:    #keep exact only once edge
    Gnr.remove_edge(G_edge_list[i][0],G_edge_list[i][1])
    
G.remove_nodes_from(list(nx.isolates(G)))  #REMOVE ISOLATE NODES
print(nx.info(G))
Gnr.remove_nodes_from(list(nx.isolates(Gnr)))  #REMOVE ISOLATE NODES
print(nx.info(Gnr))
```
